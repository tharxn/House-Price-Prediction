```python
import sys
print(sys.version)

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
```

```python
home_data = pd.read_csv("train.csv")
home_data.head()
```

```python
home_data.shape
```

```python
home_data.columns
```

```python
home_data.head()
```

```python
y = home_data.SalePrice
```

```python
features = ["LotArea","YearBuilt","1stFlrSF","2ndFlrSF","FullBath","BedroomAbvGr","TotRmsAbvGrd"]
X = home_data[features]
```

```python
train_X,val_X,train_y,val_y = train_test_split(X,y,random_state=1)
```

```python
rf_model = RandomForestRegressor(random_state=1)
rf_model.fit(train_X,train_y)
```

```python
val_predictions = rf_model.predict(val_X)
```

```python
val_mae = mean_absolute_error(val_y,val_predictions)
print("Validation MAE:", val_mae)
```

```python
rf_model_on_full_data = RandomForestRegressor(random_state=1)
rf_model_on_full_data.fit(X,y)
```

```python
test_data = pd.read_csv("test.csv")
```

```python
test_X = test_data[features]
```

```python
test_pred = rf_model_on_full_data.predict(test_X)
```

```python
output = pd.DataFrame({
    "Id" : test_data.Id,
    "SalePrice" : test_pred
})

output.to_csv("submission.csv",index=False)
output.head()
```

```python
z = np.log(home_data.SalePrice)
```

```python
W = home_data[features]
```

```python
train_W,val_W,train_z,val_z = train_test_split(W,z,random_state=1)
```

```python
rfrmse_model = RandomForestRegressor(random_state=1)
rfrmse_model.fit(train_W,train_z)
```

```python
val_preds_log = rfrmse_model.predict(val_W)
rmse = np.sqrt(mean_squared_error(val_z,val_preds_log))
print("Validation RMSe(log):",rmse)
```

```python
rfrmse_model_on_full_data = RandomForestRegressor(random_state=1)
rfrmse_model_on_full_data.fit(W,z)
```

```python
t_data = pd.read_csv("test.csv")
t_data.head()
```

```python
test_W = t_data[features] 
```

```python
t_preds_log = rfrmse_model_on_full_data.predict(test_W)
t_preds = np.exp(t_preds_log)
```

```python
sub = pd.DataFrame({
    "Id": t_data.Id,
    "SalePrice" : t_preds
})

sub.to_csv("sub.csv",index = False)
sub.head()
```
